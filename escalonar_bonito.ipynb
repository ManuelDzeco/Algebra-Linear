```python
import sympy as sp
```

```python
# Ativa a impressão bonita (frações reais, símbolos elegantes)
sp.init_printing(use_unicode = True)
```

```python
def mostrar_passo(titulo, matriz, operacoes = None):
    print(f"\n>>> {titulo.upper()} <<<")
    if operacoes:
        for op in operacoes:
            print(f"\t[Operação]: {op}")
    sp.pprint(matriz)
    print("-" * 50)
```

```python
def escalonar_bonito(matriz):
    M = sp.Matrix(matriz)
    linhas, colunas = M.shape

    mostrar_passo("Matriz Inicial", M)

    pivo_lin = 0
    for j in range(colunas):
        if pivo_lin >= linhas:
            break

        # 1. Busca de pivô e troca de linha (Pivotamento)
        if M[pivo_lin, j] == 0:
            for i in range(pivo_lin + 1, linhas):
                if M[i, j] != 0:
                    M.swapr(pivo_lin, i)
                    mostrar_passo(f"Troca de Linhas", M, [f"L{pivo_lin + 1} <-> L{i+1}"])
                    break

        if M[pivo_lin, j] != 0:
            # 2. Normalização do Pivô (Transformar em 1)
            if M[pivo_lin, j] != 1:
                fator = sp.simplify(1/M[pivo_lin, j])
                M.row_op(pivo_lin, lambda v, _: sp.simplify(v * fator))
                mostrar_passo(f"Normalização do Pivô na Coluna {j+1}", M, [f"L{pivo_lin + 1} = L{pivo_lin + 1} * ({fator})"])

            # 3. Zerar a coluna (Eliminação)
            ops_realizadas = []
            for i in range(linhas):
                if i != pivo_lin and M[i, j] != 0:
                    fator = sp.simplify(-M[i, j])
                    ops_realizadas.append(f"L{i+1} = L{i+1} + ({fator}) * L{pivo_lin+1}")
                    M.row_op(i, lambda v, k: sp.simplify(v + fator * M[pivo_lin, k]))

            if ops_realizadas:
                mostrar_passo(f"Eliminação na Coluna {j+1}", M, ops_realizadas)

            pivo_lin += 1

    print("\n" + "="*20)
    print(f"POSTO DA MATRIZ: {M.rank()}")
    print("="*20)
    return M
```

```python
lamb = sp.Symbol("λ")
matriz = [
    [3, lamb, 1, 2],
    [1, 4, 7, 2],
    [1, 10, 17, 4],
    [4, 1, 3, 3]
]
```

```python
escalonar_bonito(matriz)
```
