```python
import sympy as sp
```

Função escalonar_passo_a_passo

Receber uma matriz e a armazenar na variavel M
extrair o numero de linhas e colunas da matriz em M e armazenar nas variaveis linhas e colunas

Usar a função pprint(pretty_print) do sympy para imprimir bonito
Escrever 30 "-" para criar separação
criar a variavel pivo_lin que recebe 0

para cada f no intervalo de colunas
    se pivo_lin for maior ou igual ao numero de linhas
        fim do programa(break)

    Se o elemento M[pivo_lin(ou seja, i), j] for igual a zero
        para cada i no intervalo [pivo_lin+1,linhas]
            se M[i,j] for diferente de 0
                Trocar pivo_lin por i nas respectivas colunas
                Escrever("Passo: Troca de linha pivo_lin+i <-> i+1")
                Imprimir a matriz equavalite bonito usando a função pprint
                Fim da repetição
    Se o elemento M[pivo_lin, j] for diferente de zero
        #Agora a intenção é normalizar tal elemento, de modo que ele se torne em 1
        Se M[pivo_lin, j] for diferente de 1
            criar a variavel fator que recebe 1/M[pivo_lin, j]
            

```python
def escalonar_passo_a_passo(matriz):
    M = sp.Matrix(matriz)
    linhas, colunas = M.shape

    print("=== MATRIZ INICIAL ===")
    sp.pprint(M)
    print("-" * 30)

    pivo_lin = 0
    for j in range(colunas):
        if pivo_lin >= linhas:
            break

        # 1. Encontrar o melhor pivô na coluna atual (i)
        # Se o elemento for 0, tenta trocar com uma linha abaixo
        if M[pivo_lin, j] == 0:
            for i in range(pivo_lin + 1, linhas):
                if M[i, j] != 0:
                    M.swapr(pivo_lin, i)
                    print(f"Passo: Troca de Linha L{pivo_lin+1} <-> L{i+1}")
                    sp.pprint(M)
                    break
        if M[pivo_lin, j] != 0:
            # 2. Normalizar o pivô para 1
            if M[pivo_lin, j] != 1:
                fator = 1/ M[pivo_lin, j]
                M.row_op(pivo_lin, lambda v, _: v*fator)
                print(f"Passo: Normalizar L{pivo_lin+1} (L{pivo_lin+1} * {fator})")
                sp.pprint(M)

            # 3. Zerar outros elementos na mesma coluna (cima e baixo)
            operacoes = []
            for i in range(linhas):
                if i != pivo_lin and M[i, j] != 0:
                    fator = -M[i, j]
                    operacoes.append(f"L{i+1} = L{i+1} + ({fator})*L{pivo_lin+1}")
                    M.row_op(i, lambda v, k: v + fator * M[pivo_lin, k])

            if operacoes:
                print(f"Passo de Elimininação na Coluna {j+1}:")
                for op in operacoes:
                    print(f"\t-> {op}")
                sp.pprint(M)
                print("-" * 30)

            pivo_lin += 1

    print("\n=== RESULTADO FINAL ===")
    sp.pprint(M)
    print(f"Posto (Rank): {M.rank()}")
    return M
```

```python
lamb = sp.Symbol("λ")
matriz = [
    [3, lamb, 1, 2],
    [1, 4, 7, 2],
    [1, 10, 17, 4],
    [4, 1, 3, 3]
]
```

```python
escalonar_passo_a_passo(matriz)
```
