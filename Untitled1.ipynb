```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider
```

```python
# 1. Definir a nossa Matriz (O nosso "Mapa" ou "Sistema")
A = np.array([[2, -1],
              [1, 1]])

# 2. Configurar a janela e os dois painéis (Colunas à esquerda, Linhas à direita)
fig, (ax_col, ax_lin) = plt.subplots(1, 2, figsize = (14, 6))
plt.subplots_adjust(bottom = 0.3) # Dar espaço em baixo para os botões/sliders

# --- FUNÇÃO DE DESENHO ---
def atualizar_graficos(val):
    # Limpar os gráficos antigos
    ax_col.clear()
    ax_lin.clear()

    # Obter os valores atuais dos sliders (o nosso vetor x)
    x1 = slider_x1.val
    x2 = slider_x2.val
    x = np.array([x1, x2])

    # Calcular o vetor resultante b = A*x
    b = A @ x
    # ==========================================
    # PAINEL 1: VISÃO DAS COLUNAS (O Construtor)
    # ==========================================
    ax_col.set_title("Visão das Colunas: Combinação Linear")
    ax_col.grid(True, linestyle = '--', alpha = 0.6)
    ax_col.set_xlim(-10, 10)
    ax_col.set_ylim(-10, 10)
    ax_col.axvline(0, color = 'black', linewidth = 1)
    ax_col.axhline(0, color = 'black', linewidth = 1)

    col1 = A[:, 0]
    col2 = A[:, 1]

    # Escalar as colunas usando os valores x1 e x2
    v1_escalado = x1 * col1
    v2_escalado = x2 * col2

    # Desenhar os passos (Regra do paralelogramo/ponta-a-cauda)
    ax_col.quiver(0, 0, v1_escalado[0], v1_escalado[1], angles = 'xy', scale_units = 'xy', scale = 1, color = 'blue', label = f'{x1:.1f} * Coluna 1')
    ax_col.quiver(v1_escalado[0], v1_escalado[1], v2_escalado[0], v2_escalado[1], angles = 'xy', scale_units = 'xy', scale = 1, color = 'red', label = f'{x2:.1f} * Coluna 2')
    ax_col.quiver(0, 0, b[0], b[1], angles = 'xy', scale_units = 'xy', scale = 1, color = 'green', width = 0.015, label = f'Resultado b: {b}')
    ax_col.legend(loc = 'upper left')

    # PAINEL 2: VISÃO DAS LINHAS (O Detective)
    # ==========================================
    ax_lin.set_title("Visão das Linhas: Interseção")
    ax_lin.grid(True, linestyle = '--', alpha = 0.6)
    ax_lin.set_xlim(-5, 5)
    ax_lin.set_ylim(-5, 5)
    ax_lin.axvline(0, color = 'black', linewidth = 1)
    ax_lin.axhline(0, color = 'black', linewidth = 1)

    # Gerar pontos X para desenhar as retas
    x_vals = np.linspace(-5, 5, 100)

    # A matemática das linhas: a11*x + a12*y = b1 => y = (b1 - a11*x) / a12
    # Linha 1: 2x - 1y = b[0] -> y = 2x - b[0]
    y1_vals = 2 * x_vals - b[0]
    # Linha 2: 1x + 1y = b[1] -> y = -x + b[1]
    y2_vals = -x_vals + b[1]

    ax_lin.plot(x_vals, y1_vals, color = 'purple', label = f'Eq 1: 2x - y = {b[0]:.1f}')
    ax_lin.plot(x_vals, y2_vals, color = 'orange', label = f'Eq 2: x + y = {b[1]:.1f}')

    # Onde elas se cruzam? Extamente no nosso vetor de entrada (x1, x2)!
    ax_lin.plot(x1, x2, 'ko', markersize = 8, label = f'Interseção: ({x1:.1f},{x2:1f})')
    ax_lin.legend(loc = 'upper left')

# --- CONFIGURAR OS SLIDERS (A interface de utilizador) ---
ax_slider_x1 = plt.axes([0.25, 0.15, 0.50, 0.03])
ax_slider_x2 = plt.axes([0.25, 0.08, 0.50, 0.03])

slider_x1 = Slider(ax_slider_x1, 'Valor de x1', -5.0, 5.0, valinit = 2.0)
slider_x2 = Slider(ax_slider_x2, 'Valor de x2', -5.0, 5.0, valinit = 3.0)

# Dizer aos sliders para correrem a nossa função sempre que se mexerem
slider_x1.on_changed(atualizar_graficos)
slider_x2.on_changed(atualizar_graficos)

# Desenhar o estado inicial
atualizar_graficos(None)

plt.show()
```
