```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Definiar Matriz A e Vetor x
A = np.array([[2, -1],
             [1, 1]])
x = np.array([2, 3])
```

```python
# Extrair as colunas de A
col1 = A[:, 0]
col2 = A[:, 1]
```

```python
# Calcular os vetores escalados
v1_scaled = x[0] * col1
v2_scaled = x[1] * col2
b = v1_scaled + v2_scaled # O resultado final
```

```python
# Configuração do gráfico
plt.figure(figsize=(8, 8))
plt.axvline(x = 0, color = 'grey', lw = 1)
plt.axhline(y = 0, color = 'grey', lw = 1)
plt.grid(True, linestyle = '--', alpha = 0.6)

# Desenhar as colunas originais
plt.quiver(0, 0, col1[0], col1[1], angles='xy', scale_units = 'xy', scale = 1, color = 'blue', alpha = 0.3, label = 'Coluna 1 de A')
plt.quiver(0, 0, col2[0], col2[1], angles='xy', scale_units = 'xy', scale = 1, color = 'red', alpha = 0.3, label = 'Coluna 2 de A')

# Desenhar a combinação linear(vetores escalados que formam o vetor final)
plt.quiver(0, 0, v1_scaled[0], v1_scaled[1], angles = 'xy', scale_units='xy', scale = 1, color = 'blue', label = f'{x[0]} * Coluna 1')
plt.quiver(v1_scaled[0], v1_scaled[1], v2_scaled[0], v2_scaled[1], angles = 'xy', scale_units = 'xy', scale = 1, color = 'red', label = f'{x[1]} * Coluna 2')

# Desenhar o vetor resultante b
plt.quiver(0, 0, b[0], b[1], angles = 'xy', scale_units = 'xy', scale = 1, color = 'green', width = 0.015, label = 'Vetor Resultante (b)')

plt.xlim(-5, 6)
plt.ylim(-2, 7)
plt.legend()
plt.title("Perspectiva da Colunas: Multiplicação de Matriz como Combinação Linear")
plt.show()
```

```python
# Perspectiva das linhas
import numpy as np
import matplotlib.pyplot as plt

# Definir as equações (2x - y = 1) e (x + y = 5)
# Vamos resolver as equações em ordem a 'y' para as poder desenhar:
# Equação 1: y = 2x - 1
# Equação 2: y = -x + 5

x_vals = np.linspace(-2, 5, 400)
y1_vals = 2 * x_vals - 1
y2_vals = -x_vals + 5

# Resolver o sistema usando a álgebra do numpy para encontrar o ponto de interseção exato
A = np.array([[2, -1], [1, 1]])
b = np.array([1, 5])
solucao_x = np.linalg.solve(A, b)

plt.figure(figsize = (8, 8))
plt.axvline(x = 0, color = 'grey', lw = 1)
plt.axhline(y = 0, color = 'grey', lw = 1)
plt.grid(True, linestyle = '--', alpha = 0.6)

# Desenhar as linhas
plt.plot(x_vals, y1_vals, color = 'purple', label = 'Linha 1: 2x - y = 1')
plt.plot(x_vals, y2_vals, color = 'orange', label = 'Linha 2: x + y = 5')

# Marcar o ponto de interseção
plt.plot(solucao_x[0], solucao_x[1], 'ko', markersize = 8, label = f'Interseção (Vetor x): ({solucao_x[0]:.0f}, {solucao_x[1]:.0f})')

plt.xlim(-2, 5)
plt.ylim(-2, 7)
plt.legend()
plt.title("Perspectiva das linhas: Interseção de Retas")
plt.show()
```
